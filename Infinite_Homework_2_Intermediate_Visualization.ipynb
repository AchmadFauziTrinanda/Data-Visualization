**Content**

Each row represents a customer, each column contains customer’s attributes described on the column Metadata.

The data set includes information about:

- Customers who left within the last month – the column is called Churn
- Services that each customer has signed up for – phone, multiple lines, internet, online security, online backup, device protection, tech support, and streaming TV and movies
- Customer account information – how long they’ve been a customer, contract, payment method, paperless billing, monthly charges, and total charges
- Demographic info about customers – gender, age range, and if they have partners and dependents

**Import Library**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
%matplotlib inline

sns.set_style('whitegrid')
plt.rc('patch', edgecolor='black')
sns.set_context("notebook", font_scale=1.3, rc={"lines.linewidth": 1.5})
```

**Read Dataset**

```python
data = pd.read_csv("https://raw.githubusercontent.com/AchmadFauziTrinanda/Datasets/main/Telco-Customer-Churn.csv")
data.head()
```

**Visualization**

```python
# Average monthly charges by gender using
internet_service_monthly = pd.DataFrame(data.groupby('InternetService').mean()['MonthlyCharges'])
internet_service_monthly
```

```python
# Sorting monthly charges
internet_service_monthly.sort_values(by=['MonthlyCharges'], inplace=True, ascending=False)
internet_service_monthly
```

```python
# Data visualization of average monthly charges by gender
plt.figure(figsize=(15, 10)) # plt.figure -> create a new figure  
internet_service_monthly.plot(kind='bar', color='royalblue') # pandas uses plot(kind='bar') to make chart/diagram/graph
plt.title('Average Internet Services by Monthly Charges', fontsize=14) # plt.title is used to give title of the chart/diagram/graph 
plt.xlabel('Internet Services', fontsize=14) # plt.xlabel is used to set the label for the x-axis
plt.ylabel('Monthly Charges', fontsize=14) # plt.ylabel is used to set the label for the y-axis
plt.grid(color='darkgray', linestyle=':', linewidth=0.5) # plt.grid is used to configure the grid lines
plt.legend(loc='right', bbox_to_anchor=(1.5, 0.5), shadow=True, ncol=0.5) # plt.legend is used to place a legend on the axes
plt.show() # plt.show is used to display all open figures
```

Based on the information above, we know that the average monthly cost for fiber optic Internet services is the most expensive average cost. After Fiber Optic, the average monthly cost of DSL internet services ranks second. And the rest are customers who do not use Internet Services.
