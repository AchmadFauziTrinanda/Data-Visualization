#Make Visualization

**Bar Chart**

```python
#Perbandingan konsumsi coal 1980 dengan 2012
data1.sort_values(by='2012',ascending=False)[:20].plot.bar(x='Area', y=['1980','2012'])
```

The coal consumption is higher in 2012 than the coal consumption in 1980. The majority of coal consumption is seggregated in Asia Oceania (China & India) and North America (United States)

**Histogram**

```python
data1['1980'].plot.hist(subplots=True, layout = (1,1))
```

```python
data1['2012'].plot.hist()
```

The histogram shows how the coal consumption distributed. The majority of countries consume less than 10 quadrillion Btu.

**KDE Distribution**

```python
data1['1980'].plot.kde(subplots=True)
data1['2012'].plot.kde(subplots=True, color = 'red')
```

The KDE diagram shows the same thing with the histogram where majority of the countries in the world consume lesser than China and US

**Scatter Diagram**

```python
data1.sort_values(by='1980',ascending=False)[:50].plot.scatter(x='1980', y='Area', title='1980 Consumption',fontsize = 8,figsize = (10,10),subplots=True,layout = (1,1))
data1.sort_values(by='2012',ascending=False)[:50].plot.scatter(x='2012', y='Area', title='2012 Consumption',fontsize = 8,figsize = (10,10),subplots=True,layout = (1,1))
```

This scatter shows how the increment of coal consumption from Asia & Oceania region which is the contribution of China's coal consumption

**Pie Chart**

```python
tertinggi = data1.sort_values(by='1980',ascending=False)[2:10]
tertinggi.groupby(['Area']).sum().plot(kind='pie', y='1980', legend= False,autopct="%.2f",figsize = (5,5),subplots=True,layout = (1,1))
tertinggi.groupby(['Area']).sum().plot(kind='pie', y='2012', legend= False,autopct="%.2f",figsize = (5,5),subplots=True,layout = (1,1))
```

The proportion of coal consumption from the year 1980 and 2012. It shows how china dominates the coal consumption from 13.15% to 53.60% of the total consumption of the top global coal consumers. This consumption also affects the economic growth of the country. 

**Box Plot**

```python
data1.sort_values(by='2012',ascending=False)[2:234].plot.box(figsize=(30,10))
```

The boxplot shows many outliers which is the indsutrial countries which consume more coal than the others.

**Line & Area Chart**

```python
data1_line = data1.set_index("Area").sort_values(by='2012',ascending=False)[:10].T
ax = data1_line.plot.line(title="Coal Consumption")
ax.set_title("Coal Consumption",color='black')
ax.legend(bbox_to_anchor=(1.0, 1.0))
ax.plot()
```

```python
ax = data1_line.plot.area(stacked=False);
ax.set_title("Coal Consumption",color='black')
ax.legend(bbox_to_anchor=(1.0, 1.0))
ax.plot()
```

Both graph show how the increment of coal consumption in the world from 1980 to 2012

```python
data_scatter = data1_line.reset_index()
data_scatter['index']=pd.to_numeric(data_scatter.index,errors='coerce')
ax=data_scatter.plot.line(x='index', y=['World','China'])
```

and when we focus on this graph, we can see that World's coal consumption growth has the same trend with China's consumption. It indicates the China's contribution to the total World's Consumption

**Stacked Bar Chart**

```python
ax = data1_line.plot.barh(stacked=True,figsize=(10,10));
ax.set_ylim(21.5,32.5)
ax.set_title("Coal Consumption (2002-2012",color='black')
ax.plot()
```

This graph show the comparison and proportion of coal consumption from all area in the last 10 years (2002-2012)
